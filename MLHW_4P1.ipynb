```python
#Seth Hensley
#801055921
#homework 4.1

import warnings
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

from sklearn import metrics
from sklearn import datasets 
from sklearn.datasets import load_breast_cancer
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix 
from sklearn.metrics import classification_report
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn import svm
```

```python
cancerData = load_breast_cancer()
cancerX = pd.DataFrame(cancerData.data)
cancerY = pd.DataFrame(cancerData.target)

#cancerX,cancerY

sc = StandardScaler()
cancerXSTD = sc.fit_transform(cancerX)
cancerXSTD#cancerY.shape
```

```python
N= 25
for i in range(1,25):
    print('principle component(s)=' ,i)
    pca = PCA(n_components = i)
    principalComponents = pca.fit_transform(cancerXSTD)
    principalDF = pd.DataFrame(data = principalComponents)

    principalDF

    xTrain,xTest,yTrain,yTest = train_test_split(principalDF,cancerY,test_size = .2, random_state = 20)

    C = [.1]
    for c in C:
        clf = LogisticRegression(penalty = 'l1', C=c, solver = 'liblinear')
        clf.fit(xTrain, yTrain)
        print('c:', c)
        print( 'Training accuracy:', clf.score(xTrain,yTrain))
        print('test accuracy:', clf.score(xTest,yTest))
        print('')
    warnings.filterwarnings('ignore')


    predicted = clf.predict(xTest)
    matrix = confusion_matrix(yTest,predicted)
    print(matrix)


    report = classification_report(yTest,predicted)
    print(report)
```

```python
#it appears that after the number of principle componenets exceeds 5 the result from regression training do not improve
# 5 principle components will be used 
```

```python
pca = PCA(n_components = 5)
principalComponents = pca.fit_transform(cancerXSTD)
principalDF = pd.DataFrame(data = principalComponents)

xTrain,xTest,yTrain,yTest = train_test_split(principalDF,cancerY,test_size = .2, random_state = 20)
```

```python
clf = svm.SVC(kernel = 'linear',C=1E6)
clf.fit(xTrain,yTrain)
predictedSVM = clf.predict(xTest)

cfLinear = confusion_matrix(yTest,predictedSVM)
reportLinear = classification_report(yTest,predictedSVM)
print(cfLinear)
print(reportLinear)
print('test accuracy', metrics.accuracy_score(yTest,predictedSVM))
```

```python
#for 5 principle components there is not much of a diffrence in 
#precision recall or accuracy. for lower numbers of principle components 
#the precision and recall are improved however the overall
#test accuracy stays about the same. 
#for a lower amount of principle components a linear kernel worked best
```

```python
clf = svm.SVC(kernel = 'poly',C=1E6)
clf.fit(xTrain,yTrain)
predictedSVM = clf.predict(xTest)

cfLinear = confusion_matrix(yTest,predictedSVM)
reportLinear = classification_report(yTest,predictedSVM)
print(cfLinear)
print(reportLinear)
print('test accuracy', metrics.accuracy_score(yTest,predictedSVM))
```
